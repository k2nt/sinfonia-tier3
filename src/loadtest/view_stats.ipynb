```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
REPORT_ROOT_FOLDER = Path("./report")
REPORT_ID = "1709716699"
SESSION_ROOT_FOLDER = str(Path(REPORT_ROOT_FOLDER) / REPORT_ID)
```

```python
rdf50 = pd.read_csv(Path(SESSION_ROOT_FOLDER) / "carbon-report-50.0rps-8u.csv")
rdf400 = pd.read_csv(Path(SESSION_ROOT_FOLDER) / "carbon-report-400.0rps-8u.csv")
rdf800 = pd.read_csv(Path(SESSION_ROOT_FOLDER) / "carbon-report-800.0rps-8u.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming report_df is your Pandas DataFrame with the specified fields
# If not, replace it with your actual DataFrame

x50 = rdf50['timestamp']
x50 = x50 - x50.min()

x400 = rdf400['timestamp']
x400 = x400 - x400.min()

x800 = rdf800['timestamp']
x800 = x800 - x800.min()

field = 'energy_use_joules'
y50 = rdf50[field]
y400 = rdf400[field]
y800 = rdf800[field]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x50, y50, linestyle='-', linewidth=1, label='50')
plt.plot(x400, y400, linestyle='--', linewidth=1, label='400')
plt.plot(x800, y800, linestyle='-.', linewidth=1, label='800')

# Customize the plot
plt.title('Carbon Emission Over Time')
plt.xlabel('Seconds since start')
plt.ylabel('Carbon Emission (gCO2)')


# Show legend to differentiate between series
plt.legend()

# Rotate x-axis labels for better readability (optional)
# plt.xticks(rotation=45)

# Display the plot
plt.show()
```
