```python
import os
from pathlib import Path
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def idx(u, rps, msz) -> str:
    return f"latency-report-{u}u-{round(rps, 1)}rps-{msz}msz.csv"
```

```python
TS = "1714146202"
U = 1
MSZ = 25
RPS = [10.0, 50.0, 100.0, 150.0, 200.0]
REPORT_PATH = Path("./report") / TS
```

```python
latency_dfs = dict()
for rps in RPS:
    fn = idx(U, rps, MSZ)
    latency_dfs[fn] = pd.read_csv(REPORT_PATH / fn, names=["timestamp", "avg", "95"])
```

```python
for rps in RPS:
    df = latency_dfs[idx(U, rps, MSZ)]

    plt.plot(df['timestamp'], df['avg'], label='avg')
    plt.plot(df['timestamp'], df['95'], label='95')

    plt.xlabel('timestamp')
    plt.ylabel('Value')
    plt.title(f'rps={rps}')
    plt.legend()

    plt.show()
```
